# Striplogs from images

```python
from striplog import Striplog
from legend import Legend
legend = Legend.default()
```

```python
import os
env = %env
```

```python
directory = os.path.join(env['HOME'], "Dropbox/Agile/NS_DOE/2014-1G/data/PNG")
```

```python
import re
def listdir(directory, match):
    for f in os.listdir(directory):
        if match:
            if not re.search(match, f, flags=re.IGNORECASE):
                continue
        yield os.path.join(directory, f)
```

```python
wells = {}
for f in listdir(directory, r"[MP].+\.png"):
    well, start, stop = os.path.basename(f).strip('.png').split('_')
    wells[well] = Striplog.from_img(f, float(start), float(stop), legend=legend, tolerance=35)
```

```python
import templates
```

```python
import time
```

```python
def form_las(well, striplog):
    time_now = time.strftime("%Y/%m/%d %H:%M", time.gmtime())
    result = templates.las.format(
                  prog='striplog.py',
                  date=time_now,
                  start=striplog.start,
                  stop=striplog.stop,
                  step=-999.25,
                  null=-999.25,
                  well=well,
                  uwi='',
                  lic=well,
                  apd=-999.25,
                  eref=-999.25,
                  section=striplog.to_las3(),
                  curve='')
    return result.strip() + '\n'
```

```python
print form_las('P-129', wells['P-129'])
```

```python
def write_las(data, path, suffix=None, overwrite=False):
    if not suffix:
        suffix = time.strftime("%Y%m%d-%H%M%S", time.gmtime())
    
    w = re.findall(r'\nWELL \.(.+?)\:WELL', data)[0].strip()
    file_name = re.sub(r' ', r'_', w) + '_{0}.las'.format(suffix)
    
    full_path = os.path.join(path, file_name)
    
    if (not os.path.isfile(full_path)) or overwrite:
        with open(full_path, 'w') as f:
            f.write(data)
    else:
        print w, "... file exists, not overwritten"
```

```python
data = form_las('P-129', wells['P-129'])
path = os.path.join(env['HOME'], "Dropbox/Agile/NS_DOE/2014-1G/data/LAS")
write_las(data, path, 'striplog', overwrite=True)
```
